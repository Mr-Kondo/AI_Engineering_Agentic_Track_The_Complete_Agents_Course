```python
# 1. 設定済みのクライアントを作成
import os
from agents import set_tracing_export_api_key, Agent, Runner, trace, RunConfig
from agents.extensions.models.litellm_model import LitellmModel
from dotenv import load_dotenv
load_dotenv(override=True)
tracing_api_key = os.environ["OPENAI_API_KEY"]
set_tracing_export_api_key(tracing_api_key)

model = LitellmModel(
    base_url='http://localhost:11434/v1',
    api_key="ollama",
    model = 'ollama/llama3.1:8b',
)

# 2. クライアントをAgentに渡す
# (注: ライブラリによって引数名が 'client' だったり 'openai_client' だったりします)
agent = Agent(
    name="Jokester",
    instructions="You are a joke teller",
    model='llama3.1:8b',
)
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents", run_config=RunConfig(tracing={"api_key": "sk-tracing-123"}))
    print(result.final_output)
```
